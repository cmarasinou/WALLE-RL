```python
import numpy as np
import pandas as pd
from gridworld import Environment, Agent
import matplotlib.pyplot as plt
```

```python
def update_utility(utility, eligibility_trace, alpha, gamma, state_prime, state, reward_prime):
    delta = reward_prime + gamma*utility[state_prime]-utility[state]
    utility = utility + alpha*delta*eligibility_trace
    return utility
    
def update_eligibilty_trace(eligibility_trace, gamma,lambd, state):
    eligibility_trace_prime = gamma*lambd*eligibility_trace
    eligibility_trace_prime[state] =  gamma*lambd*eligibility_trace[state]+1
    return eligibility_trace_prime
```

```python
def TD_lambda_episode(utility, eligibility_trace, alpha=0.1, gamma=0.9, lambd = 0.5):
    ''' Runs the temporal difference algorithm for one episode
    Args:
        utility (1d array, (n_states)): the initial utility
        eligibility_trace (1d array, (n_states)): the initial eligibility trace
        alpha (float, optional): learning_rate
        gamma (float, optional): discount rate
        lamb (float, optional): eligibilty trace decay rate
    Returns:
        utility (1d array, (n_states)): resulting utility
    '''
    agent = Agent()
    env = Environment()
    state = env.current_state
    reward = env.initial_reward
    episode_finished=False
    # for tracking utility
    df = pd.DataFrame([[0]+list(utility)],columns=['iter','(0, 0)',
                             '(0, 1)',
                             '(0, 2)',
                             '(0, 3)',
                             '(1, 0)',
                             'x',
                             '(1, 2)',
                             '-1',
                             '(2, 0)',
                             '(2, 1)',
                             '(2, 2)',
                             '1'])
    i = 0
    while not episode_finished:
        i += 1
        action = agent.step(state)
        state_prime, reward_prime, episode_finished = env.step(action)
        eligibility_trace = update_eligibilty_trace(eligibility_trace, gamma,lambd, state)
        utility = update_utility(utility, eligibility_trace, alpha,gamma,state_prime, state, reward_prime)
        state, reward = state_prime, reward_prime
        df.loc[i+1] = [int(i+1)]+list(utility)
    return utility, df
```

```python
# initializing
utility = np.zeros(12)
eligibility_trace = np.zeros(12)
lambd = 0.0
utility, df = TD_lambda_episode(utility, eligibility_trace, lambd = lambd)
plt.figure(dpi = 100)
# style
plt.style.use('seaborn-whitegrid')
 
# create a color palette
palette = plt.get_cmap('tab20')
 
# multiple line plot
num=0
for column in df.drop('iter', axis=1):
    num+=1
    plt.plot(df['iter'], df[column], marker='', color=palette(num), linewidth=1.5, alpha=0.9, label=column)

# Add legend
plt.legend(loc=1, ncol=2, fontsize=7)
 
# Add titles
plt.title("TD lambda = {}".format(lambd), fontsize=15, fontweight=0, color='black')
plt.xlabel("Episodes")
plt.ylabel("Utility")
#plt.ylim(-0.1,0.1)
plt.show()
```
